# 5. 데이터 전처리

## 5.1 기간에서 연도를 분리하기

```python
 int("2019 4/4 p)".split()[0]) # 년도 분기/월
```

```python
df['연도']= list(map(lambda x : int(x.split(".")[0]), df["기간"]))
df.head()
```

```python
"2019 4/4 p)".split()[1].split("/")[0]
```

```python
df['분기']=list(map(lambda x: int(x.split(".")[1].split("/")[0]), df['기간']))
df.head()
```

## 5.2 금액을 수치데이터로 표현하기 위해 데이터 타입 변경하기

```python
#- 문자를 결측지로 변경하여 float타입으로 만들기
df['백만원']=df['백만원'].replace('-', pd.np.nan).astype(float)
df.head()
```

## 5,3 필요없는 데이터 삭제하기

```python
df.info()
```

```python
df = df[(df["국가(대륙)별"] != "합계") & (df["상품군별"] != "합계")].copy()
```

```python
df.info()
```

```python
df.isnull().sum()
```
