# Summerize Data

```python
# 데이터프레임
df = pd.DataFrame(
{"a" : [4, 5, 6, 4],
"b" : [7, 8, 9, 10],
"c" : [10, 11, 12, 13]},
index = [1, 2, 3, 4])
df
```

```python
# 카테고리컬한 데이터의 갯수를 샐 때 사용
df[["a"]].value_counts()
```

```python
len(df)
```

# Reshaping


sort_values, drop


```python
# 특정값을 기준으로 정렬
df.sort_values("a",ascending=False)
```

```python
# drop 기본 axis =0
b_df=df.drop(["a","c"],axis=1)
```

```python
# drop 기본 axis =0
df=df.drop(["c"],axis=1)
df
```

# Group Data

Groupby, pivot_table

```python
df.groupby(["a"])["b"].agg(["mean","sum","count"])
```

```python
df.groupby(["a"])["b"].describe()
```

```python
# pivot은 형태만 바꾸고 pivot_table은 연산도 할 수 있는 차이
# 기본값이 mean
pd.pivot_table(df, index="b", values="a", aggfunc="sum")
```
