# 여러 파일 handling하기


```python
# 여러 파일을 다루는 방법
# 데이터 - 소상공인시장진흥공단_상가(상권)정보 : https://www.data.go.kr/data/15083033/fileData.do
```

```python
# 파일 목록 가져오기
file_lst=glob("/content/*_*_*.csv")
file_lst
```

```python
files=[]
for file_name in file_lst:
  print(file_name)
  temp = pd.read_csv(file_name, encoding ='utf-8',low_memory=False)
  files.append(temp)
```
